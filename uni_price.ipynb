```python
table_211 = []
soup = BeautifulSoup(html, 'html.parser')

this_table = soup.find('table')

trs = this_table.find_all('tr')

for tr in trs[1:]:
    Year = []
    for td in tr.find_all('td'):
        Year.append(td.text)
    table_211.append(Year)

table_final =[]

for Year in table_211:
    dict= {}
    dict['Year'] = Year[0]
    dict['Landings (million lbs)'] = Year[1]
    dict['Average ex-vessel price'] = Year[2]
    table_final.append(dict)
    
table_final

    
    

    
```

```python
import pandas as pd
df = pd.DataFrame(table_final)
df.head()
```

```python
df['Landings (million lbs)'] = df['Landings (million lbs)'].astype(float)
df['Average ex-vessel price'] = df['Average ex-vessel price'].str.replace('$', '').astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Landing change in %'] = df['Landings (million lbs)'].pct_change()*100
df['Average ex-vessel price change in %'] = df['Average ex-vessel price'].pct_change()*100
df.head()
```

```python
df.to_csv('uni_price.csv', index=False)
```
